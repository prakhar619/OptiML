```python
def MSE(w,x = x_new,y = y_):
    y_hat = x@w
    loss = np.sum(np.square(y-y_hat))
    return loss
```

```python
# w is row/col vector
def der(w,x,y_):
    P  =  x@w
    delta_p_L = 2 * (P - y_)
    delta_n_L = delta_p_L
    delta_w_L_alt = x_new.T@delta_n_L
    delta_w_L_alt
    return delta_w_L_alt
```

### Perceptron

```python
import sklearn.datasets as sk
import numpy as np
```

```python
x_1, y_1 = sk.make_blobs(n_samples = 1000, centers = 2, n_features = 2,random_state = 0,cluster_std=0.5)
```

```python
x_1
```

```python
import matplotlib.pyplot as plt
```

```python
colors = y_1*100
plt.scatter(x_1[:,0],x_1[:,1], c=colors, cmap='viridis')
```

```python
y_1 = (y_1 * 2) - 1
y_1
```

```python
w = np.random.rand(2)
b = np.random.rand()
```

```python
def train(w,x,y):
    for i in range(len(x)):
        y_hat_i = 1 if (x[i]@w.T) + b > 0 else -1 
        if y_hat_i != y[i]:
            w = w + y[i]*x[i] 
    return w
def predict(w,x):
    y_hat = []
    for i in range(len(x)):
        y_hat_i = 1 if (x[i]@w.T) + b > 0 else -1
        y_hat.append(y_hat_i)
    return y_hat
```

```python
print(w)
w = train(w,x_1,y_1)
predict(w,x_1)
```

```python
colors = y_1*100
plt.scatter(x_1[:,0],x_1[:,1], c=colors, cmap='viridis')
plt.plot((w),[0,b], c='red', label='line')
```
