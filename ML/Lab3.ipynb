### Q1

```python
iid = np.random.rand(100)
iid
```

```python
def bernouli_model(p,li):
    dataset = []
    for i in li:
        if i < p:
            dataset.append(1)
        else:
            dataset.append(0)
    return dataset
```

```python
dataset = bernouli_model(0.4,iid)
dataset
```

```python
def likelihood(dataset,prob):
    overall_prob = 1
    for i in dataset:
        if i == 1:
            overall_prob *= prob
        else:
            overall_prob *= (1-prob)
    return overall_prob
```

```python
y = []
x= np.arange(0,1.05,0.05)
for p in x:
    y.append(likelihood(dataset,p))
```

```python
import matplotlib.pyplot as plt
```

```python
peak = max(y)
peak_index = y.index(peak)
x_cor = x[peak_index]
plt.axvline(x = x_cor, color='r')
plt.plot(x,y)
```

```python
def logLikelihood(dataset,prob):
    x = likelihood(dataset,prob)
    return np.log(x)
```

```python
y = []
x= np.arange(0,1.05,0.05)
for p in x:
    y.append(logLikelihood(dataset,p))
```

```python
peak = max(y)
peak_index = y.index(peak)
x_cor = x[peak_index]
plt.axvline(x = x_cor, color='r')
plt.plot(x,y)
```

```python
def loss_h1(dataset):
    loss = 0
    for i in dataset:
        if i == 0:
            loss +=1
    return loss
```

```python
def loss_h0(dataset):
    loss = 0
    for i in dataset:
        if i == 1:
            loss +=1
    return loss
```

```python
def loss_hz(dataset,prob):
    loss = 0
    for i in dataset:
        r = np.random.rand()
        if r < prob:
            predict = 1
        else:  
            predict = 0
        if predict != i:
            loss +=1 
    return loss
```

```python
print(loss_h0(dataset))
print(loss_h1(dataset))
print(loss_hz(dataset,p))
```

```python
y = []
x= np.arange(0,1.05,0.01)
for p in x:
    y.append(loss_hz(dataset,p))
```

```python
plt.axhline(y = loss_h0(dataset), color='y')
plt.axhline(y = loss_h1(dataset), color = 'r')
plt.plot(x,y)
```
