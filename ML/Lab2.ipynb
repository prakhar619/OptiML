### Perceptron

```python
import sklearn.datasets as sk
import numpy as np
```

```python
x_1, y_1 = sk.make_blobs(n_samples = 1000, centers = 2, n_features = 2,random_state = 0,cluster_std=0.5)
```

```python
x_1
```

```python
import matplotlib.pyplot as plt
```

```python
colors = y_1*100
plt.scatter(x_1[:,0],x_1[:,1], c=colors, cmap='viridis')
```

```python
y_1 = (y_1 * 2) - 1
y_1
```

```python
w = np.random.rand(2)
b = np.random.rand()
```

```python
def train(w,x,y):
    for i in range(len(x)):
        y_hat_i = 1 if (x[i]@w.T) + b > 0 else -1 
        if y_hat_i != y[i]:
            w = w + y[i]*x[i] 
    return w
def predict(w,x):
    y_hat = []
    for i in range(len(x)):
        y_hat_i = 1 if (x[i]@w.T) + b > 0 else -1
        y_hat.append(y_hat_i)
    return np.array(y_hat)
```

```python
epoch = 1000
for i in range(epoch):
    w = train(w,x_1,y_1)
y_hat = predict(w,x_1)
y_hat
```

```python
colors = (y_hat+1)*30
plt.scatter(x_1[:,0],x_1[:,1], c=colors, cmap='plasma')

x_bnd = np.linspace(x_1[:, 0].min() - 1, x_1[:, 0].max() + 1,  500)
y_bnd = - x_bnd * (w[0] /w[1])
plt.plot(x_bnd, y_bnd,color='red')

plt.plot()
```
