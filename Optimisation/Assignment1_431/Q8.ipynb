```python
import numpy as np 
import matplotlib.pyplot as plt
import math 
from mpl_toolkits.mplot3d import axes3d
```

```python
def rosenbrock(x1,x2):
    return (1- x1)**2 + 100*(x2 - x1**2)**2
```

```python
def d_rosenbrock(x1,x2):
    d_x1 = 2*(1- x1)*(-1) + 100*2*(x2 - x1**2) *(-2*x1)
    d_x2 = 100*2*(x2 - x1**2)
    return (d_x1,d_x2)
```

```python
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)
```

```python
fig, ax = plt.subplots()
cs = ax.contour(X,Y,rosenbrock(X,Y))
manual_location = [0.5,2,5,10,50,100,200,400,800]
ax.clabel(cs,inline=True,fontsize=10,manual=manual_location)
```
