```python
def forward_layer_matrixMult(x, w):
    # both w,x are row vectors
    # x : n * d, w : d * 1
    # N = xw

    return x @ w
```

```python
def foraward_layer_biasAdd(N, b):
    # N : n * 1, b : n * 1
    # P = N + b

    return N + b
```

```python
def forward_layer_sigmoid(P):
    # Q = 1 / 1 + e^-P
    
    return 1 / (1 + np.exp(-P))
```

```python
def forward_layer_meanSqrLoss(P,y):
    # P : n * 1, y : n * 1
    # L = 1/n * sum((P - y) ^ 2)    : (1 * 1)
    
    return np.sum((P - y)**2)
```
