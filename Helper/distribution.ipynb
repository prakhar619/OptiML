```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
count = 100000
norm = np.random.normal(0,10,size=(count))
plt.hist(norm,bins=1000)
count
```

```python
const = 20
norm1 = const + norm
plt.hist(norm1,bins=1000)
count
```

```python
alpha = 20
norm2 = norm * alpha
plt.hist(norm2,bins=1000)
count
```
