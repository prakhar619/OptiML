```python
import numpy as np
```

```python
#Parameters
N = 1000
k = 5
n = 10
p = np.random.rand(k)
p
```

```python
pi = np.random.rand(k)
sum = np.sum(pi)
pi = pi/sum
pi = np.round(pi,decimals=2)
pi,np.sum(pi)
```

```python

datasample = []
for i in range(N):
    coin_c = np.random.randint(k)
    p_head = p[coin_c]
    sample_i = []
    for i in range(n):
        r = np.random.rand()
        if( r <= p_head):
            sample_i.append(1)
        else:
            sample_i.append(0)
    datasample.append(sample_i)
```

```python
datasample
```

```python
from math import comb
def E(p_approx,pi_approx):
    gamma = np.zeros((N,k))
    for i in range(0,N):
        xi = np.sum(datasample[i])
        sum = 0
        for c in range(0,k):
            gamma[i][c] = pi_approx[c] * comb(n,xi) * (p_approx[c]**xi) * ((1-p_approx[c])**(n-xi))
            sum += gamma[i][c]
        for c in range(0,k):
            gamma[i][c] = gamma[i][c]/sum
    return gamma
```

```python
def M(gamma):   
    p_approx = np.zeros(k)
    pi_approx = np.zeros(k)
    for j in range(k):
        sum = 0
        for i in range(N):
            sum += gamma[i][j]
        pi_approx[j] = sum/N
    for r in range(k):
        sum = 0
        denom = 0
        for i in range(N):
            xi = np.sum(datasample[i])
            sum += gamma[i][r] * xi
            denom += gamma[i][r] * n
            p_approx[r] = sum/denom
    return p_approx,pi_approx
```

```python
#Approximate Parameter
p_approx = np.random.rand(k)
pi_approx = np.random.rand(k)
pi_approx = pi_approx / np.sum(pi_approx)
pi_approx
```

```python
def converge(p_approx,pi_approx):
    for i in range(1000):
        gamma = E(p_approx,pi_approx)
        p_old=  p_approx.copy()
        pi_old = pi_approx.copy()
        p_approx,pi_approx = M(gamma)
        if( np.sum((p_old - p_approx)**2) < 1e-5 and np.sum((pi_old - pi_approx)**2) < 1e-5):
            return p_approx,pi_approx
        
```

```python
p_new,pi_new = converge(p_approx,pi_approx)
```
