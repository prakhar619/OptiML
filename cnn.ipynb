```python
import numpy as np
```

```python
#Parameters
inputdim = (28,28)
inputCount = 1
alpha = 0.01
kernel_size = (5,3)
```

```python
x = np.random.randint(255,size=(inputCount,inputdim[0],inputdim[1]))
```

```python
def print_matrix(x):
    print(x.shape)
    a = 0
    for k in x:
        print("Sample:",a)
        a+=1
        for i in k:
            for j in i:
                print(j,end=",\t")
            print()
        print()
```

```python
print_matrix(x)
```

```python
import math as Math
pad_width = ((0,0),(Math.floor(kernel_size[0]/2),Math.floor(kernel_size[0]/2)),(Math.floor(kernel_size[1]/2),Math.floor(kernel_size[1]/2)))
padded_x =  np.pad(x, pad_width, mode='constant', constant_values=0)
```

```python
print_matrix(padded_x)
```

```python
kernel = np.random.rand(kernel_size[0],kernel_size[1])
kernel
```

```python
#Forward Pass
output_dim = inputdim
y_hat = np.zeros((inputCount,output_dim[0],output_dim[1]))
```

```python
for sample_count in range(inputCount):
    for i in range(output_dim[0]):
        for j in range(output_dim[1]):
            y_hat[sample_count,i,j] = np.sum(kernel.T@padded_x[sample_count,i:i+kernel_size[0], j:j+kernel_size[1]] )
```

```python
print_matrix(y_hat)
```

```python
def der(f_Rn_R,w):
    diagonal_elements = np.random.rand(len(w)) * 1e-5
    diagonal_matrix = np.diag(diagonal_elements) + w                            #SQUARE MATRIX
    f_x_h = np.apply_along_axis(f_Rn_R, axis=1, arr=diagonal_matrix)            #ROW MATRIX
    f_x = np.apply_along_axis(f_Rn_R, axis=1,arr = np.tile(w,(len(w),1)))       #ROW MATRIX
    delta_w_L = ( f_x_h - f_x ) / diagonal_elements                             #ROW MATRIX
    return delta_w_L
```
